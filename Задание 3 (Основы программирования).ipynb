# Основы программирования

1. Дан массив целых чисел nums и целое число target. Написать функцию, возвращающую индексы элементов, дающих в сумме число target?
2. Предполагается, что в каждом массиве имеется не больше одной пары дающих в сумме заданное число target. Нельзя использовать один и тот же элемент дважды?


```python
nums = [2, 5, 6, 1, 4]
target = 3
```

Первый вариант - перебрать через двойной цикл все возможные комбинации элементов

```python
for i in range(len(nums)):
  n_i = nums[i]
  for j in range(len(nums)):
    n_j = nums[j]
    if (n_i + n_j) == target:
      print (i, j)
      break
```

Получается, что мы перебираем n*n = n^2 комбинаций, где n - длина массива nums

Обратим внимание, что мы можем изменить диапазоны циклов, чтобы не рассматривать суммы первых и последних элементов сами с собой

```python
for i in range(0,len(nums)-1):
  n_i = nums[i]
  for j in range(1,len(nums)):
    n_j = nums[j]
    if (n_i + n_j) == target:
      print (i, j)
      break
```

Число итераций сократилось до (n-1)(n-1) = n^2-2n+1, то есть на 2n-1

Попробуем использовать индексы и то что не более однлой пары:

```python
[nums.index(target-num) for num in nums if (target-num) in nums]
```

```python
set([[nums.index(num), nums.index(target-num)] for num in nums if (target-num) in nums])
```

Таким методом мы проходимся по списку лишь раз, что значительно сокращает импользуемые время и память

Можно доработать данный алгоритм, чтобы индексы были в порядке возрастания

```python
[nums.index(num) for num in nums if (target-num) in nums]
```

Создадим пользовательскую функцию

```python
def task1 (nums, target):
  return([nums.index(num) for num in nums if (target-num) in nums])
```
